```python
import pandas as pd
```

```python

data_original=pd.read_csv("002 flight-price.csv")
data=data_original.copy()
```

```python
data.info()
data.describe()
```

```python
data.isnull().sum()
```

```python
data[data.duplicated()]
```

```python
data.drop_duplicates(inplace=True)
```

```python
data["Dep_Time"].str.isnumeric().sum()
data["Arrival_Time"].str.isnumeric().sum()
```

```python
data["Dep_Time"].unique()
```

```python
data["Date_of_Journey"]=pd.to_datetime(data["Date_of_Journey"])
```

```python
data["Date_of_Journey"]
```

```python
data["Arrival_Time"].unique()
```

```python
data["Arrival_Time Hours"]=data["Arrival_Time"].str.split(':').str[0]
data["Arrival_Time Minutes"]=data["Arrival_Time"].str.split(':').str[1]
data["Arrival_Time Minutes"]=data["Arrival_Time Minutes"].str.split(' ').str[0]
```

```python
data
```

```python
data["Arrival_Time Hours"]=data["Arrival_Time Hours"].astype(int)
data["Arrival_Time Minutes"]=data["Arrival_Time Minutes"].astype(int)
```

```python
cols=list(data.columns)
cols.remove('Arrival_Time Hours')
cols.remove('Arrival_Time Minutes')

pos=cols.index('Arrival_Time')
cols.insert(pos+1,'Arrival_Time Hours')
cols.insert(pos+2,'Arrival_Time Minutes')
data=data[cols]
```

```python
data
```

```python
data.drop("Arrival_Time",inplace=True,axis=1)
```

```python
data
```

```python
data.info()
```

```python
data["Duration"].unique()
```

```python
data["Duration"]=data["Duration"].fillna(0)
data["Duration"].unique()
```

```python
data["Duration Hours"]=data["Duration"].str.split(" ").str[0]
data["Duration Minutes"]=data["Duration"].str.split(" ").str[1]
```

```python
data
```

```python
data["Duration Hours"]=data["Duration Hours"].str.replace("h","")
data["Duration Minutes"]=data["Duration Minutes"].str.replace("m"," ")
```

```python
data["Duration Minutes"]=data["Duration Minutes"].fillna(0)
```

```python
data["Duration Hours"]=data["Duration Hours"].fillna(0)
```

```python
data.loc[6474,"Duration Minutes"]=5
data.loc[6474,"Duration Hours"]=0
```

```python
data["Duration Minutes"]=data["Duration Minutes"].astype(int)
data["Duration Hours"]=data["Duration Hours"].astype(int)
```

```python
cols=list(data.columns)
cols.remove('Duration Hours')
cols.remove('Duration Minutes')

pos=cols.index('Duration')
cols.insert(pos+1,'Duration Hours')
cols.insert(pos+2,'Duration Minutes')
data=data[cols]
```

```python
data
```

```python
data.info()
```

```python
data["Total_Stops"]=data["Total_Stops"].str.replace("non-stop","0")
```

```python
data['Total_Stops']=data['Total_Stops'].str.split(" ").str[0]
```

```python
data["Total_Stops"]=data["Total_Stops"].fillna(0)
```

```python
data['Total_Stops']=data['Total_Stops'].astype(int)
```

```python
data.info()
```

```python
data["Date_of_Journey"]
```

```python
data["Date_of_Journey"]=data["Date_of_Journey"].astype(str)
```

```python
data['Year']=data['Date_of_Journey'].str.split("-").str[0]
data['Month']=data['Date_of_Journey'].str.split("-").str[1]
data['Date']=data['Date_of_Journey'].str.split("-").str[2]
```

```python
cols=list(data.columns)
cols.remove('Year')
cols.remove('Month')
cols.remove('Date')

pos=cols.index('Date_of_Journey')
cols.insert(pos+1,'Date')
cols.insert(pos+2,'Month')
cols.insert(pos+3,'Year')
data=data[cols]
```

```python
data
```

```python
data['Date']=data['Date'].astype(int)
data['Month']=data['Month'].astype(int)
data['Year']=data['Year'].astype(int)
```

```python
data["Dep_Time Hours"]=data["Dep_Time"].str.split(":").str[0]
data["Dep_Time Minutes"]=data["Dep_Time"].str.split(":").str[1]
```

```python
cols=list(data.columns)
cols.remove('Dep_Time Hours')
cols.remove('Dep_Time Minutes')

pos=cols.index('Dep_Time')
cols.insert(pos+1,'Dep_Time Hours')
cols.insert(pos+2,'Dep_Time Minutes')
data=data[cols]
```

```python
data["Dep_Time Hours"]=data["Dep_Time Hours"].astype(int)
data["Dep_Time Minutes"]=data["Dep_Time Minutes"].astype(int)
```

```python
data
```

```python
data.drop(columns={"Date_of_Journey","Duration","Dep_Time"},inplace=True)
```

```python
data
```

```python
data.info()    
```

```python
data.to_csv("data/cleanedfully.csv",index=False)
```
